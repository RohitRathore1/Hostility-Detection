```
def check_tweet_exists(test_tweet, texts_list,percent):
    counter = 1
    test_tweet     = test_tweet.split()

    for existing_tweet in texts_list:
        existing_tweet = existing_tweet.split()
        n              = len(set(existing_tweet))
        m              = 0

        if(len(test_tweet)>n and len(test_tweet)/n > 10):
            return False

        matching_words = []

        for word in test_tweet:
            if word in existing_tweet:
                # m = m + 1
                if(word not in matching_words):
                    matching_words.append(word)
                    m = m + 1

                if(m/n > percent):    
                    print()
                    print(existing_tweet)
                    print(test_tweet)
                    print() 
                    return True
    return False
```

# Cleaning Rohit's Tweets

```
df = pd.read_csv("/gdrive/My Drive/IIITD Fake News Project/Data Set/OLD/Hindi Tweets Annotation Rohit.csv")
df
```

```
texts_list = []
percent = 0.90

drop_indexes = []
counter = 0

for text in df['Tweet Text']:
    if(check_tweet_exists(text,texts_list,percent)):
        drop_indexes.append(counter)

    texts_list.append(text)
    counter = counter + 1
    if(counter>2500):
        break
    


print("The Number of tweets with more than ",percent*100,"% similarity is: ",len(drop_indexes))
```

```
print(len(texts_list))
print(len(drop_indexes))
```

```
df = df.drop(index = drop_indexes)
df
```
