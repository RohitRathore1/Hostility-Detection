```
# url definition
url = 'https://hindi.boomlive.in/fake-news/'

r1 = requests.get(url)
r1.status_code

# We'll save in coverpage the cover page content
coverpage = r1.content

# Soup creation
soup1 = BeautifulSoup(coverpage, 'html5lib')

# News identification
coverpage_news = soup1.find_all('h2', class_='entry-title')
len(coverpage_news)
```

```
list_links = []
for n in np.arange(1,12):
  # Getting the link of the article
  link = coverpage_news[n].find('a')['href']
  list_links.append(link)
```

```
all_divs = soup1.findAll('h2', attrs={'class':'entry-title'})
for h2 in all_divs:
  links = h2.find('a')['href']
  print(links)
```

```
list_links = []
for i in range(1,200):
  url = 'https://hindi.boomlive.in/fake-news/'
  url = url + str(i) + '?infinitescroll=1'
  res = requests.get(url)
  soup = BeautifulSoup(res.content, 'html.parser')
  all_divs = soup.findAll('h2', class_='entry-title')
  for h2 in all_divs:
    links = h2.find('a')['href']
    list_links.append(links)
```

```
# df_show_info
df_show_info = pd.DataFrame(
    {'Article Link': list_links})
```

```
all_hindi_news = []
for i in range(1,200):
  url = 'https://hindi.boomlive.in/fake-news/'
  url = url + str(i) + '?infinitescroll=1'
  res = requests.get(url)
  soup = BeautifulSoup(res.content, 'html.parser')
  all_divs = soup.findAll('div', attrs={'class':'card-content'})
  for div in all_divs:
    links = div.find('a')
    print(links.text)
    all_hindi_news.append(links.text)
```
